```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Select the columns to scale
columns_to_scale = ['price_change_percentage_24h', 'price_change_percentage_7d']
market_data_to_scale = df_market_data[columns_to_scale]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(market_data_to_scale)
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original DataFrame
# Set the coin_id column as index
df_scaled = pd.DataFrame(scaled_data, columns=columns_to_scale, index=df_market_data.index)

# Display the scaled DataFrame
df_scaled
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    KMeans_model = KMeans(n_clusters=k, random_state=1)
    KMeans_model.fit(df_scaled)
    inertia_values.append(KMeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

#displaying the DataFrame
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=700,
    height=400
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the elbow curve, the optimal value for k appears to be 3, as that’s where we observe the most noticeable drop in inertia before the curve begins to flatten. Although k=4 was a common value in past activities, in this specific dataset, k=3 provides the clearest natural grouping of the data.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
k_best = 3
kmeans_model = KMeans(n_clusters=k_best, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
crypto_clusters = kmeans_model.predict(df_scaled)

# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the scaled DataFrame
df_clusters = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_clusters["cluster"] = crypto_clusters

# Display the copy of the scaled DataFrame
df_clusters.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_clusters.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster",
    hover_cols=["coin_id"],
    title="Crypto Clusters by 24h and 7d Price Changes",
    width=700,
    height=500
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=2)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_scaled)

#Sanity Check - creating a DataFrame with the PCA data
df_pca = pd.DataFrame(
    pca_data,
    columns=["PC1", "PC2"],
    index=df_scaled.index
)

# View the scaled PCA data
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# Display the explained variance
explained_variance
```

```python
# Calculate total explained variance
total_explained_variance = explained_variance.sum()
print(f"Total explained variance: {total_explained_variance:.4f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The original dataset only contained two features, so PCA could only reduce the data to two principal components. As a result, the total explained variance is the sum of the two components, which is approximately 1.0000. This still captures 100% of the variance in the original dataset, just spread across two axes instead of three.

```python
# Create a new DataFrame with the PCA data.
df_pca_final = df_pca.copy()

# Copy the crypto names from the original scaled DataFrame
df_pca_final["coin_id"] = df_scaled.index

# Set the coin_id column as index
df_pca_final = df_pca_final.set_index("coin_id")

# Display the scaled PCA DataFrame
df_pca_final.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
#renaming a Dataframe so it stops erroring out
df_market_data_pca = df_pca_final
```

```python
# Create a list with the number of k-values from 1 to 11
k_values_pca = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values_pca:
    kmeans_pca = KMeans(n_clusters=k, random_state=1)
    kmeans_pca.fit(df_market_data_pca)
    inertia_values_pca.append(kmeans_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    "k": k_values_pca,
    "inertia": inertia_values_pca
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)

#displaying the DataFrame
df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow_pca.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve Using PCA Data",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=700,
    height=400
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k when using the PCA data is 3, as that’s where the inertia sharply decreases before the curve begins to level off. This suggests three distinct clusters capture most of the structure in the PCA-reduced data.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ. Both the original scaled data and the PCA-reduced data point to k=3 as the optimal number of clusters, indicating that PCA preserved the structure of the original dataset while reducing dimensionality.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
k_best_pca = 3
kmeans_pca_final = KMeans(n_clusters=k_best_pca, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca_final.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_clusters = kmeans_pca_final.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_clusters = df_market_data_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_clusters["cluster"] = pca_clusters

# Display the copy of the scaled PCA DataFrame
df_pca_clusters.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_pca_clusters.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="cluster",
    hover_cols=["coin_id"],
    title="Crypto Clusters Using PCA Data",
    width=700,
    height=500
)
```
